```python
# importing necessary libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup # this function is used to parse and work with the HTML we fetched from our Wiki page:
import requests
```

```python
# getting data from internet
data = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text

# using beautiful soup to parse the HTML/XML codes.
a = BeautifulSoup(data,'xml')
print(a.prettify())
```

```python
# Based on HTML markers we are trying to find the tabel of our requierment - the class of table that we were able to search from above is wikitable sortable
right_tab = a.find('table', class_='wikitable sortable')
right_tab
```

```python
# writing a loop to go through the HTML code and fill in the corrosponding cells

Postcode      = []
Borough       = []
Neighbourhood = []

for row in right_tab.findAll('tr'):
        cell = row.findAll('td')
        if len(cell)==3:
            Postcode.append(cell[0].find(text=True))
            Borough.append(cell[1].find(text=True))
            Neighbourhood.append(cell[2].find(text=True))
```

```python
# converting the list into data frame
df = pd.DataFrame(Postcode, columns=['Postcode'])
df['Borough'] = Borough
df['Neighbourhood'] = Neighbourhood
df.head()
```

```python
# creating a new data frame without borough having Not assigned
df2 = df[df.Borough.str.contains('Not assigned') == False].reset_index(drop=True)
df2.head()
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
Toronto_venues = getNearbyVenues(names=df4['Neighbourhood'],
                                   latitudes = df4['Latitude'],
                                   longitudes=df4['Longitude']
                                  )
```

```python
Toronto_venues.head()
```

```python
# one hot encoding
Toronto_onehot = pd.get_dummies(Toronto_venues[['Venue Category']], prefix="", prefix_sep="" )
# add neighborhood column back to dataframe
Toronto_onehot['Neighbourhood'] = Toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [Toronto_onehot.columns[-1]] + list(Toronto_onehot.columns[:-1])
Toronto_onehot = Toronto_onehot[fixed_columns]

Toronto_onehot.head()
```

```python
Toronto_grouped = Toronto_onehot.groupby('Neighbourhood').mean().reset_index()
Toronto_grouped .head()
```

```python
from sklearn.cluster import KMeans 
Toronto_grouped_clustering = Toronto_grouped.drop('Neighbourhood', 1)

kmeans = KMeans(init='k-means++', n_clusters=5, n_init = 12)
kmeans.fit(Toronto_grouped_clustering)
Toronto_grouped_clustering.head()
```

```python
kmeans.labels_[0:10] 
```

```python
# Let's create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

# add clustering labels

Toronto_grouped.insert(0, 'Cluster Labels', kmeans.labels_)

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
Toronto_merged = df_toronto.join(Toronto_grouped.set_index('Neighbourhood'), on='Neighbourhood')

Toronto_merged.head()
```
